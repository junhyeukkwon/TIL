# 퍼셉트론
* 프랑크 로젠블라트가 1957년에 고안한 알고리즘이다. 
* 이 알고리즘이 딥러닝이라고 불리는 신경망의 기원이 되는 알고리즘이기 때문에 중요하다. 
* 퍼셉트론 신호는 흐른다/안흐른다 (1 or 0)의 두 가지 값만 가질 수 있다. 


* $ 𝑥_1 $  과 $ 𝑥_2$ 는 입력 신호 
* $ 𝑤_1 $와 $𝑤_2 $는 가중치를 의미 
* 그림의 원은 뉴런 혹은 노드
* 입력 신호가 뉴런에 보내질 때는 각각 고유한 가중치가 곱해진다.($ 𝑤_1 𝑥_1 ,𝑤_2  𝑥_2$)
* 뉴런에서 보내온 신호의 총합이 정해진 한계를 넘어설 때만 1을 출력한다. 이를 뉴런이 활성화한다라는 표현을 하고 그 한계를 임계값이라며 하며, 𝜃라 한다. 


* 복수의 입력 신호 각각에 고유한 가중치를 부여하고, 가중치는 각 신호가 결과에 주는 영향력을 조절하는 요소로 작용한다. 즉 가중치가 클수록 신호가 그만큼 더 중요함을 의미한다. 


# 단순한 논리회로
* 컴퓨터를 구성하고 있는 논리회르를 퍼셉트론을 활용하면 간단하게 구현할 수 있습니다.

## AND 게이트
* AND 게이트는 두 입력이 모두 1일 때만 1을 출력하고, 그 외에는 0을 출력합니다. 
* 아래는 AND 게이트의 진리표입니다. 

```python
import numpy as np
```

## AND Gate
* 퍼셉트론의 구조를 사용해서 $ w_1, w_2, \theta $ 값만 지정하면 쉽게 구현할 수 있습니다. 

```python
def AND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.7
    tmp = np.sum(w*x) + b
    if tmp <= 0:
        return 0
    else:
        return 1
```

```python
for xs in [(0, 0), (1, 0), (0, 1), (1, 1)]:
    y = AND(xs[0], xs[1])
    print(str(xs) + " -> " + str(y))
```

## OR Gate
* OR 게이트는 두 값이 둘 다 0일때만 0을 출력하고 그외에는 모두 1을 출력합니다. 

* AND와 동일하게 퍼셉트론 구조로 OR를 구현할 수 있습니다. 

```python
def OR(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.2
    tmp = np.sum(w*x) + b
    if tmp <= 0:
        return 0
    else:
        return 1
```

```python
for xs in [(0, 0), (1, 0), (0, 1), (1, 1)]:
    y = OR(xs[0], xs[1])
    print(str(xs) + " -> " + str(y))
```

## NAND Gate
* NAND 게이트는 Not AND를 의미하며, 그 동작은 AND 게이트의 값을 반대로 출력합니다. 
* NAND 게이트도 퍼셉트론 구현할 수 있습니다 

```python
def NAND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([-0.5, -0.5])
    b = 0.7
    tmp = np.sum(w*x) + b
    if tmp <= 0:
        return 0
    else:
        return 1
```

```python
for xs in [(0, 0), (1, 0), (0, 1), (1, 1)]:
    y = NAND(xs[0], xs[1])
    print(str(xs) + " -> " + str(y))
```

## 다층 퍼셉트론
* 퍼셉트론을 하나만 사용하는 것이 아니라 여러 층으로 쌓아서 사용하는 다층 퍼셉트론(multi-layer perceptron)을 만들 수 있습니다. 
* 기존의 AND, OR, NAND 선형 게이트를 사용해서 XOR를 만들어 보겠습니다. 

```python
def XOR(x1, x2):
    s1 = NAND(x1, x2)
    s2 = OR(x1, x2)
    y = AND(s1, s2)
    return y
```

```python
for xs in [(0, 0), (1, 0), (0, 1), (1, 1)]:
    y = XOR(xs[0], xs[1])
    print(str(xs) + " -> " + str(y))
```

* XOR은 아래와 같은 다층 구조의 네트워크입니다. 
* 왼쪽부터 0층, 1층, 2층이라고 부르겠습니다. 
* 이처럼 층이 여러 개인 퍼셉트론을 다층 퍼셉트론이라고 부릅니다. 

* 층을 쌓으면 단층 퍼셉트론으로는 표현하지 못한 것을 표현할 수 있습니다.

## 활성화 함수의 종류 
### 시그모이드 함수 
* 신경망에서 자주 이용하는 활성화 함수로는 시그모이드함수(sigmoid function)이 있습니다. 
* 로지스틱회귀 함수에서도 나오는 함수입니다. 

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
X = np.arange(-5, 5, 0.1)
```

```python
Y = sigmoid(X)
```

```python
plt.plot(X, Y )
plt.ylim(-.1, 1.1)
plt.show()
```

## 계단함수
* 입력이 0을 넘으면 1을 출력하고, 그 외에는 0을 출력하는 함수입니다. 

```python
def step_function(x):
    return np.array(x > 0, dtype=int)
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y = step_function(x)
```

```python
plt.plot(x,y)
plt.ylim(-0.1, 1.1)
plt.show()
```

## ReLU 함수
* ReLU(Rectified Linear) 함수는 최근에 많이 사용하는 함수입니다. 
* 입력이 0을 넘으면 그 입력을 그대로 출력하고, 0 이하면 0을 출력하는 함수입니다. 

```python
def relu(x):
    return np.maximum(0, x)
```

```python
x = np.arange(-5, 5, 0.1)
y = relu(x)
```

```python
plt.plot(x, y)
plt.show()
```

# 신경망의 연산(내적)
* 신경망은 행렬 연산을 사용하여 계산합니다. 
* 고차원의 행렬을 텐서(Tensor)라고 부릅니다. 텐서플로우는 고차원 행렬 연산을 쉽게 해주는 프레임윅입니다. 

* 참고) 텐서는 다양한 차원을 가질 수 있는데 위의 그림은 3차원 텐서를 나타낸 것입니다. 3차원 텐서는 3차 텐서(3th order tensor)라고 하며, 일반적으로 n차원 텐서는 n차 텐서(nth-order tensor)라고 부릅니다.
* 벡터는 1차원 텐서라고 부를 수 있습니다. 
* 행렬은 2차원 텐서에 해당하므로 2차 텐서(second-order tensor)라고 부릅니다. 
* 3차 이상부터는 고차원 텐서(higher-order tensor)라고 부릅니다. 

## 행렬의 내적

```python
X = np.array([1., 0.5])
W1 = np.array([[0.1, 0.3, 0.5], [0.2, 0.4, 0.6]])
B1 = np.array([0.1, 0.2, 0.3])
```

```python
A1 = np.dot(X, W1) + B1
```

```python
Z1 = sigmoid(A1)
```

```python
W2 = np.array([[0.1, 0.4], [0.2, 0.5], [0.3, 0.6]])
B2 = np.array([0.1, 0.2])
```

```python
Z2 = sigmoid(np.dot(Z1, W2) + B2)
```

```python
W3 = np.array([[0.1,0.3], [0.2, 0.4]])
B3 = np.array([0.1, 0.2])
```

```python
def identity_function(x):
    return x
```

```python
Y = identity_function(np.dot(Z2, W3) + B3)
```

```python
def softmax(x):
    if x.ndim == 2:
        x = x.T
        x = x - np.max(x, axis=0)
        y = np.exp(x) / np.sum(np.exp(x), axis=0)
        return y.T 

    x = x - np.max(x) # 오버플로 대책
    return np.exp(x) / np.sum(np.exp(x))
```
