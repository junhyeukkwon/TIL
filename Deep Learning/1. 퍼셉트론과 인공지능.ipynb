# Îã®ÏàúÌïú ÎÖºÎ¶¨ÌöåÎ°ú

## AND Í≤åÏù¥Ìä∏
* AND Í≤åÏù¥Ìä∏Îäî Îëê ÏûÖÎ†•Ïù¥ Î™®Îëê 1Ïùº ÎïåÎßå 1ÏùÑ Ï∂úÎ†•ÌïòÍ≥†, Í∑∏ Ïô∏ÏóêÎäî 0ÏùÑ Ï∂úÎ†•Ìï©ÎãàÎã§. 
* ÏïÑÎûòÎäî AND Í≤åÏù¥Ìä∏Ïùò ÏßÑÎ¶¨ÌëúÏûÖÎãàÎã§. 

```python
import numpy as np
```

## AND Gate
* ÌçºÏÖâÌä∏Î°†Ïùò Íµ¨Ï°∞Î•º ÏÇ¨Ïö©Ìï¥ÏÑú $ w_1, w_2, \theta $ Í∞íÎßå ÏßÄÏ†ïÌïòÎ©¥ ÏâΩÍ≤å Íµ¨ÌòÑÌï† Ïàò ÏûàÏäµÎãàÎã§. 

```python
def AND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.7
    tmp = np.sum(w*x) + b
    if tmp <= 0:
        return 0
    else:
        return 1
```

```python
for xs in [(0, 0), (1, 0), (0, 1), (1, 1)]:
    y = AND(xs[0], xs[1])
    print(str(xs) + " -> " + str(y))
```

## OR Gate
* OR Í≤åÏù¥Ìä∏Îäî Îëê Í∞íÏù¥ Îëò Îã§ 0ÏùºÎïåÎßå 0ÏùÑ Ï∂úÎ†•ÌïòÍ≥† Í∑∏Ïô∏ÏóêÎäî Î™®Îëê 1ÏùÑ Ï∂úÎ†•Ìï©ÎãàÎã§. 

* ANDÏôÄ ÎèôÏùºÌïòÍ≤å ÌçºÏÖâÌä∏Î°† Íµ¨Ï°∞Î°ú ORÎ•º Íµ¨ÌòÑÌï† Ïàò ÏûàÏäµÎãàÎã§. 

```python
def OR(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.2
    tmp = np.sum(w*x) + b
    if tmp <= 0:
        return 0
    else:
        return 1
```

```python
for xs in [(0, 0), (1, 0), (0, 1), (1, 1)]:
    y = OR(xs[0], xs[1])
    print(str(xs) + " -> " + str(y))
```

## NAND Gate
* NAND Í≤åÏù¥Ìä∏Îäî Not ANDÎ•º ÏùòÎØ∏ÌïòÎ©∞, Í∑∏ ÎèôÏûëÏùÄ AND Í≤åÏù¥Ìä∏Ïùò Í∞íÏùÑ Î∞òÎåÄÎ°ú Ï∂úÎ†•Ìï©ÎãàÎã§. 
* NAND Í≤åÏù¥Ìä∏ÎèÑ ÌçºÏÖâÌä∏Î°† Íµ¨ÌòÑÌï† Ïàò ÏûàÏäµÎãàÎã§ 

```python
def NAND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([-0.5, -0.5])
    b = 0.7
    tmp = np.sum(w*x) + b
    if tmp <= 0:
        return 0
    else:
        return 1
```

```python
for xs in [(0, 0), (1, 0), (0, 1), (1, 1)]:
    y = NAND(xs[0], xs[1])
    print(str(xs) + " -> " + str(y))
```

## Îã§Ï∏µ ÌçºÏÖâÌä∏Î°†
* ÌçºÏÖâÌä∏Î°†ÏùÑ ÌïòÎÇòÎßå ÏÇ¨Ïö©ÌïòÎäî Í≤ÉÏù¥ ÏïÑÎãàÎùº Ïó¨Îü¨ Ï∏µÏúºÎ°ú ÏåìÏïÑÏÑú ÏÇ¨Ïö©ÌïòÎäî Îã§Ï∏µ ÌçºÏÖâÌä∏Î°†(multi-layer perceptron)ÏùÑ ÎßåÎì§ Ïàò ÏûàÏäµÎãàÎã§. 
* Í∏∞Ï°¥Ïùò AND, OR, NAND ÏÑ†Ìòï Í≤åÏù¥Ìä∏Î•º ÏÇ¨Ïö©Ìï¥ÏÑú XORÎ•º ÎßåÎì§Ïñ¥ Î≥¥Í≤†ÏäµÎãàÎã§. 

```python
def XOR(x1, x2):
    s1 = NAND(x1, x2)
    s2 = OR(x1, x2)
    y = AND(s1, s2)
    return y
```

```python
for xs in [(0, 0), (1, 0), (0, 1), (1, 1)]:
    y = XOR(xs[0], xs[1])
    print(str(xs) + " -> " + str(y))
```

# ÌçºÏÖâÌä∏Î°†ÏóêÏÑú Ïã†Í≤ΩÎßùÏúºÎ°ú

# ÌôúÏÑ±Ìôî Ìï®Ïàò

* ÏÑ†ÌòïÏùÑ ÎπÑÏÑ†ÌòïÏúºÎ°ú ÎßåÎì§Í∏∞ ÏúÑÌï¥ ÏÇ¨Ïö©ÌïúÎã§.

* ÏûÖÎ†• Ïã†Ìò∏Ïùò Ï¥ùÌï©Ïù¥ h(x)ÎùºÎäî Ìï®ÏàòÎ•º Í±∞Ï≥ê Î≥ÄÌôòÎêòÏñ¥, Í∑∏ Î≥ÄÌôòÎêú Í∞íÏù¥ yÏùò Ï∂úÎ†•Ïù¥ Îê©ÎãàÎã§. 
* ÏûÖÎ†• Ïã†Ìò∏Ïùò Ï¥ùÌï©ÏùÑ Ï∂úÎ†• Ïã†Ìò∏Î°ú Î≥ÄÌôòÌïòÎäî Ìï®ÏàòÎ•º ÏùºÎ∞òÏ†ÅÏúºÎ°ú ÌôúÏÑ±Ìôî Ìï®Ïàò(activation function)Ïù¥ÎùºÍ≥† Ìï©ÎãàÎã§. 
* ÌôúÏÑ±ÌôîÎùºÎäî Ïù¥Î¶ÑÏ≤òÎüº ÏûÖÎ†• Ïã†Ìò∏Ïùò Ï¥ùÌï©Ïù¥ ÌôúÏÑ±ÌôîÎ•º ÏùºÏúºÌÇ§ÎäîÏßÄ Ï†ïÌïòÎäî Ïó≠Ìï†ÏùÑ Ìï¥Ï§çÎãàÎã§. 
* ÌôúÏÑ±Ìôî Ìï®ÏàòÎ•º Ìè¨Ìï®ÌïòÏó¨ Í∑∏Î¶ºÏùÑ Îã§Ïãú Í∑∏Î†§Î≥¥Î©¥ ÏïÑÎûòÏôÄ Í∞ôÏäµÎãàÎã§.
* ÌïôÏäµÌïúÎã§Îäî Í≤ÉÏùÄ Îç∞Ïù¥ÌÑ∞Ïóê ÎßûÍ≤å w1, w2, bÎ•º Ï∞æÏïÑÍ∞ÄÎäîÍ≤ÉÏù¥Îã§.

## ÌôúÏÑ±Ìôî Ìï®ÏàòÏùò Ï¢ÖÎ•ò 
### ÏãúÍ∑∏Î™®Ïù¥Îìú Ìï®Ïàò 
* Ïã†Í≤ΩÎßùÏóêÏÑú ÏûêÏ£º Ïù¥Ïö©ÌïòÎäî ÌôúÏÑ±Ìôî Ìï®ÏàòÎ°úÎäî ÏãúÍ∑∏Î™®Ïù¥ÎìúÌï®Ïàò(sigmoid function)Ïù¥ ÏûàÏäµÎãàÎã§. 
* Î°úÏßÄÏä§Ìã±ÌöåÍ∑Ä Ìï®ÏàòÏóêÏÑúÎèÑ ÎÇòÏò§Îäî Ìï®ÏàòÏûÖÎãàÎã§. 

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
X = np.arange(-5, 5, 0.1)
```

```python
Y = sigmoid(X)
```

```python
plt.plot(X, Y )
plt.ylim(-.1, 1.1)
plt.show()
```

## Í≥ÑÎã®Ìï®Ïàò
* ÏûÖÎ†•Ïù¥ 0ÏùÑ ÎÑòÏúºÎ©¥ 1ÏùÑ Ï∂úÎ†•ÌïòÍ≥†, Í∑∏ Ïô∏ÏóêÎäî 0ÏùÑ Ï∂úÎ†•ÌïòÎäî Ìï®ÏàòÏûÖÎãàÎã§. 

```python
def step_function(x):
    return np.array(x > 0, dtype=int)
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y = step_function(x)
```

```python
plt.plot(x,y)
plt.ylim(-0.1, 1.1)
plt.show()
```

## ReLU Ìï®Ïàò
* ReLU(Rectified Linear) Ìï®ÏàòÎäî ÏµúÍ∑ºÏóê ÎßéÏù¥ ÏÇ¨Ïö©ÌïòÎäî Ìï®ÏàòÏûÖÎãàÎã§. 
* ÏûÖÎ†•Ïù¥ 0ÏùÑ ÎÑòÏúºÎ©¥ Í∑∏ ÏûÖÎ†•ÏùÑ Í∑∏ÎåÄÎ°ú Ï∂úÎ†•ÌïòÍ≥†, 0 Ïù¥ÌïòÎ©¥ 0ÏùÑ Ï∂úÎ†•ÌïòÎäî Ìï®ÏàòÏûÖÎãàÎã§. 

```python
def relu(x):
    return np.maximum(0, x)
```

```python
x = np.arange(-5, 5, 0.1)
y = relu(x)
```

```python
plt.plot(x, y)
plt.show()
```

# Ïã†Í≤ΩÎßùÏùò Ïó∞ÏÇ∞(ÎÇ¥Ï†Å)
* Ïã†Í≤ΩÎßùÏùÄ ÌñâÎ†¨ Ïó∞ÏÇ∞ÏùÑ ÏÇ¨Ïö©ÌïòÏó¨ Í≥ÑÏÇ∞Ìï©ÎãàÎã§.

* Ï∞∏Í≥†) ÌÖêÏÑúÎäî Îã§ÏñëÌïú Ï∞®ÏõêÏùÑ Í∞ÄÏßà Ïàò ÏûàÎäîÎç∞ ÏúÑÏùò Í∑∏Î¶ºÏùÄ 3Ï∞®Ïõê ÌÖêÏÑúÎ•º ÎÇòÌÉÄÎÇ∏ Í≤ÉÏûÖÎãàÎã§. 3Ï∞®Ïõê ÌÖêÏÑúÎäî 3Ï∞® ÌÖêÏÑú(3th order tensor)ÎùºÍ≥† ÌïòÎ©∞, ÏùºÎ∞òÏ†ÅÏúºÎ°ú nÏ∞®Ïõê ÌÖêÏÑúÎäî nÏ∞® ÌÖêÏÑú(nth-order tensor)ÎùºÍ≥† Î∂ÄÎ¶ÖÎãàÎã§.
* Î≤°ÌÑ∞Îäî 1Ï∞®Ïõê ÌÖêÏÑúÎùºÍ≥† Î∂ÄÎ•º Ïàò ÏûàÏäµÎãàÎã§.

* 3Ï∞® Ïù¥ÏÉÅÎ∂ÄÌÑ∞Îäî Í≥†Ï∞®Ïõê ÌÖêÏÑú(higher-order tensor)ÎùºÍ≥† Î∂ÄÎ¶ÖÎãàÎã§. 

## ÌñâÎ†¨Ïùò ÎÇ¥Ï†Å

```python
X = np.array([1., 0.5])
W1 = np.array([[0.1, 0.3, 0.5], [0.2, 0.4, 0.6]])
B1 = np.array([0.1, 0.2, 0.3])
```

```python
A1 = np.dot(X, W1) + B1
```

```python
Z1 = sigmoid(A1)
```

```python
W2 = np.array([[0.1, 0.4], [0.2, 0.5], [0.3, 0.6]])
B2 = np.array([0.1, 0.2])
```

```python
Z2 = sigmoid(np.dot(Z1, W2) + B2)
```

```python
W3 = np.array([[0.1,0.3], [0.2, 0.4]])
B3 = np.array([0.1, 0.2])
```

```python
def identity_function(x):
    return x
```
